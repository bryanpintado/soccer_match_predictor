```python
df = pd.read_csv("matches.csv", index_col=0)
```

```python
#change data types date and time
df['date'] = pd.to_datetime(df['date'])
df['time'] = df['time'].str.replace(':.+','',regex=True).astype(int)
```

```python
#representing object categories as numerical categories for the model
df['venue_code'] = df['venue'].astype('category').cat.codes
```

```python
df['opp_code'] = df['opponent'].astype('category').cat.codes
```

```python
df['day_code'] = df['date'].dt.dayofweek
```

```python
df['result_code'] = df['result'].astype('category').cat.codes
```

```python
df['target'] = df['result'] =='W'
```

```python
df['target'] = df['target'].astype('category').cat.codes
```

```python
#df.head(5)
```

```python
#importing model
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators = 50, min_samples_split=10, random_state=1)
```

```python
#spliting train and test data
train = df[df['date'] < '2022-01-01']
test = df[df['date'] > '2022-01-01']
```

```python
predictors = ['day_code','time', 'venue_code','opp_code']
```

```python
#fit the model
rf.fit(train[predictors],train['target'])
```

```python
preds = rf.predict(test[predictors])
```

```python
 #determining accuracy of the model to determine accuracy of predictions
from sklearn.metrics import accuracy_score
```

```python
acc = accuracy_score(test["target"], preds)
```
