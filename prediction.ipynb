```python
#change data types date and time
df['date'] = pd.to_datetime(df['date'])
df['time'] = df['time'].str.replace(':.+','',regex=True).astype(int)
```

```python
#representing object categories as numerical categories for the model
df['venue_code'] = df['venue'].astype('category').cat.codes
```

```python
df['opp_code'] = df['opponent'].astype('category').cat.codes
```

```python
df['day_code'] = df['date'].dt.dayofweek
```

```python
df['result_code'] = df['result'].astype('category').cat.codes
```

```python
df['target'] = df['result'] =='W'
```

```python
df['target'] = df['target'].astype('category').cat.codes
```
